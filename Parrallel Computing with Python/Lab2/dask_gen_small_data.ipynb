### Data Generation

- Each CSV file contains a snapshot at a given time of a speed of a list of boats
- Pandas is used to generate the files.
- In this first example, we follow 20 boats for 10 time steps.
- Each time step corresponds to a file
- Cell #1 creates the initial snapshot, Cell #2 makes it evolve


```python
import random
import pandas as pd

size = 20 # Nb boats
rounds = 10 # nb of time steps

boat_ids = range(size)
boat_speeds = [random.randint(0,15) for i in range(size)]
ports = ['BREST', 'VALENCIA', 'PALERMO', 'BRIGHTON', 'AMSTERDAM', 'STOCKHOLM', 'CORK']
ports_list = [random.choice(ports) for i in range(size)]
boats_data_init = {
    'ID': boat_ids,
    'Speed': boat_speeds,
    'Destination': ports_list
}

df_boats_step0 = pd.DataFrame(boats_data_init)
```

```python

def evolve(boat_df):
    random.seed()
    boat_speeds_updated = [max(0, boat_df['Speed'][i]+random.randint(-3,3)) for i in range(size)]
    new_boat_df = {
        'ID': boat_df['ID'],
        'Speed': boat_speeds_updated,
        'Destination': boat_df['Destination']
    }
    return(new_boat_df)

dfs = []
dfs.append(df_boats_step0)

df_boat_next = df_boats_step0
for i in range(rounds):
    df_boats_next = pd.DataFrame(evolve(df_boat_next))
    dfs.append(df_boats_next)
```

```python
for i in range(rounds):
    filename = 'boats_small'+str(i)+'.csv'
    dfs[i].to_csv(filename)
```
