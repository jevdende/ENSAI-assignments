```python
import random
import pandas as pd

size = 200000 # Nb boats
rounds = 10*60 # number of rounds in 1 hour (one round every 10 sec)

PORTS = ['BREST', 'VALENCIA', 'PALERMO', 'BRIGHTON', 'AMSTERDAM', 'STOCKHOLM', 'CORK']

boat_ids = range(size)
boat_speeds = [random.randint(0,15) for i in range(size)]
boat_destinations = [random.choice(PORTS) for i in range(size)]

boat_data_init = {
    'ID': boat_ids,
    'Speed': boat_speeds,
    'Destination': boat_destinations
}

boat_df_init = pd.DataFrame(boat_data_init)
```

```python

# generating one boat_df from another one stochastically varying the speed
def evolve(boat_df):
    random.seed()
    boat_speeds_updated = [max(0, boat_df['Speed'][i]+random.randint(-3,3)) for i in range(size)]
    new_boat_df = {
        'ID': boat_df['ID'],
        'Speed': boat_speeds_updated,
        'Destination': boat_df['Destination']
    }
    return(new_boat_df)

boat_df_init.to_csv('boats0.csv')

next_boat_df = boat_df_init
for i in range(rounds):
    next_boat_df = pd.DataFrame(evolve(next_boat_df))
    next_boat_df.to_csv('boats'+str(i)+'.csv')
```
